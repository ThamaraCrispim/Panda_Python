**Análise**:Dalc - consumo diário de álcool (numérico: de 1 - muito baixo a 5 - muito alto). Nota-se que os alunos que consomem álcool em nível 1, tem as notas variando de 7,3 a 18, e mediana em 12,3, dois pontos atípicos: 5 e 6,2 e outlier 0. No nível 2 variam de 7,7 a 16,7, com mediana em 12, pontos atípicos: 6,2; 7,3; 17,3; 17,7 e 18,2 e outlier em 1,8. Nível 3 tem variação de 9,6 a 18 e mediana em 12. O nível 4 apresenta que as notas variam de 12,7 a 15,5 e outlier em 7,7. Já o nível 5 começa variar em 10 e vai até 12,3 e outlier em 16. Em geral, os resultados sugerem uma tendência de notas mais baixas associadas a níveis mais altos de consumo de álcool, especialmente nos níveis 4 e 5. A presença consistente de pontos atípicos e outliers em vários níveis destaca a importância de considerar fatores individuais e outros contextos que podem influenciar o desempenho acadêmico.

```python
estatisticas_descritivas_por = df_gp.groupby('Dalc_por')['G3_por'].describe()
print(estatisticas_descritivas_por)
```

**analise**:

Não há uma diferença clara nas notas finais entre os diferentes níveis de consumo de álcool durante os dias úteis. As medianas são relativamente semelhantes em todos os grupos.
Conclusão:
Com base na análise visual do boxplot, não parece haver uma relação forte e evidente entre o nível de consumo de álcool durante os dias úteis e o desempenho acadêmico. No entanto, essa é uma análise preliminar, e seria útil realizar testes estatísticos mais rigorosos, como a Análise de Variância (ANOVA), para confirmar se há diferenças estatisticamente significativas entre os grupos.

```python
# Agrupando o DataFrame por nível de consumo de álcool durante os dias úteis
grouped_dalc_por = df_gp.groupby('Dalc_por')

# Calculando estatísticas descritivas para a nota final G3_por em cada grupo
estatisticas_descritivas = grouped_dalc_por['G3_por'].agg(['mean', 'median', 'std'])

# Exibindo as estatísticas descritivas
print(estatisticas_descritivas)
```

**analise**: As diferenças nas médias e medianas entre os grupos de consumo de álcool durante a semana não são muito grandes, indicando que o impacto direto do consumo de álcool na nota final pode não ser significativo.
No entanto, a variabilidade mais alta (desvio padrão) nos grupos de consumo de álcool durante a semana pode indicar que, em alguns casos, o consumo de álcool pode estar associado a uma maior variabilidade nas notas finais.
A análise sugere que não há uma relação linear clara entre o consumo de álcool durante a semana e as notas finais. É possível que outros fatores também estejam influenciando as notas finais dos alunos.
